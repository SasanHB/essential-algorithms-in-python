# 2.3 Insertion Sort

Insertion sort builds the final sorted list one item at a time.

It is good for small or almost sorted lists.


```python
def insertion_sort(lst):
    '''Sort the list in place using insertion sort.

    This algorithm builds the sorted list one item at a time by
    inserting each new item into its proper place among the already
    sorted elements.
    '''
    for i in range(1, len(lst)):
        key = lst[i]
        j = i - 1
        # Move elements of lst[0..i-1], that are greater than key,
        # to one position ahead to make space for key
        while j >= 0 and lst[j] > key:
            lst[j + 1] = lst[j]
            j -= 1
        # Place key after the element just smaller than it
        lst[j + 1] = key

data = [5, 2, 4, 6, 1, 3]
insertion_sort(data)
print(data)
```

Worst case time: **O(n²)**.  
Best case (already sorted): **O(n)**.


### Why Learn Insertion Sort?

Insertion sort shines on small or nearly sorted lists.  It builds the
sorted list one item at a time, much like sorting playing cards by hand.
Because of its stability and low overhead, insertion sort is used as a
building block inside more complex algorithms like TimSort and intro‑sort.
