# 5.2 Coin Change (Greedy Version)

Greedy strategy: always pick the largest coin that fits into the remaining amount.


```python
def coin_change_greedy(amount, coins):
    '''Return a list of coins summing to amount using a greedy strategy.

    The coins list does not need to be sorted; we sort it here in
    descending order to always pick the largest denomination first.
    This approach works when the coin system is canonical (like US coins),
    but it doesn't always give an optimal solution for arbitrary coin values.
    '''
    result = []
    coins = sorted(coins, reverse=True)
    for c in coins:
        # Keep using coin c as long as it fits in the remaining amount
        while amount >= c:
            amount -= c
            result.append(c)
    return result

print(coin_change_greedy(68, [25, 10, 5, 1]))

# Example where the greedy approach is not optimal:
# With coins [4,3,1] and amount=6, greedy returns [4,1,1] (3 coins),
# but [3,3] (2 coins) is a better solution.
print(coin_change_greedy(6, [4, 3, 1]))
```

```python
# Example where greedy fails
print(coin_change_greedy(6, [4, 3, 1]))  # 4,1,1 but 3,3 is better
```
