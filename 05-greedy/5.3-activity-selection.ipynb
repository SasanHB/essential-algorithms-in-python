# 5.3 Activity Selection

We pick the maximum number of non-overlapping activities.

Greedy rule: always pick the activity that finishes earliest.


```python
def select_activities(intervals):
    '''Select the maximum number of non-overlapping intervals.

    The function sorts intervals by their end time and then greedily
    picks the next activity that starts after the current one ends.
    '''
    # Sort activities by their finish time
    intervals = sorted(intervals, key=lambda x: x[1])
    result = []
    current_end = None

    for start, end in intervals:
        # If no activity chosen yet or this one starts after the last one ends
        if current_end is None or start >= current_end:
            result.append((start, end))
            current_end = end
    return result

activities = [(1, 4), (3, 5), (0, 6), (5, 7), (8, 9), (5, 9)]
print(select_activities(activities))
```
