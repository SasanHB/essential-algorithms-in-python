# 4.1 Recursion Basics

A function is recursive when it calls itself.

We always need:

- A base case  
- A recursive case


```python
def factorial_recursive(n):
    '''Compute n! recursively.

    The base cases are 0! = 1 and 1! = 1. For n > 1 we call the function
    itself with (n-1) and multiply the result by n.
    '''
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)

def factorial_iterative(n):
    '''Compute n! iteratively using a loop.'''
    result = 1
    # Start multiplying from 2 up to n
    for i in range(2, n + 1):
        result *= i
    return result

print(factorial_recursive(5))
print(factorial_iterative(5))
```

### Why Learn Recursion?

Recursion allows a function to call itself on smaller subproblems.  It
simplifies problems that have a natural hierarchical structure, such as
traversing trees, computing factorials and dividing tasks.  It is used in
parsing expressions, exploring decision trees and working with data
structures like linked lists.
