# 3.1 Introduction to Divide and Conquer

Divide and conquer:

1. Divide the problem into smaller pieces  
2. Solve each piece  
3. Combine results

Binary search and merge sort are good examples.


### Why Learn Divide and Conquer?

Divide‑and‑conquer algorithms break a problem into smaller independent
parts, solve each part and then combine the results.  This pattern leads
to efficient algorithms like merge sort, quick sort and fast exponentiation.
It is also used in tasks such as large integer multiplication, signal
processing (Fast Fourier Transform) and matrix multiplication.

By dividing a task, we can often parallelise the work, making better use
of multi‑core processors or distributed systems.

### Try it yourself
