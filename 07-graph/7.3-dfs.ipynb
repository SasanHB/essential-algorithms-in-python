# 7.3 Depth-First Search (DFS)

DFS explores along one path as far as possible, then backtracks.


```python
def dfs_recursive(graph, start, visited=None, order=None):
    '''Perform depth‑first search recursively and return visit order.'''
    if visited is None:
        visited = set()
    if order is None:
        order = []
    visited.add(start)
    order.append(start)
    for nei in graph.get(start, []):
        if nei not in visited:
            dfs_recursive(graph, nei, visited, order)
    return order

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': [],
}

print(dfs_recursive(graph, 'A'))
```

### Why Learn Depth‑First Search?

Depth‑first search (DFS) dives deep into a branch of a graph before
backtracking.  This makes it useful for tasks like detecting cycles,
performing topological sorts and finding connected components.  DFS powers
algorithms for solving puzzles (such as Sudoku), traversing file systems
and generating mazes.  It is also used in algorithms like Tarjan’s
strongly connected components.
