# 7.1 What is a Graph?

A graph has nodes (vertices) and edges (connections).

We often use an adjacency list in Python.


```python
graph = {
    "A": ["B", "C"],
    "B": ["A", "D"],
    "C": ["A", "D"],
    "D": ["B", "C"],
}

print(graph)
```

### Why Learn Graphs?

Graphs model relationships between objects.  They are used to represent road
maps, social networks, communication links, dependency graphs and many
other systems.  Understanding graphs allows us to solve problems like
finding connections, detecting cycles and optimising routes.
