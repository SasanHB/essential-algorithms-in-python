# 7.2 Breadth-First Search (BFS)

BFS explores a graph level by level using a queue.


```python
from collections import deque

def bfs(graph, start):
    '''Perform breadth-first search and return the order of visited nodes.

    BFS explores the graph level by level using a queue.
    '''
    visited = set([start])
    order = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        order.append(node)
        for nei in graph.get(node, []):
            if nei not in visited:
                visited.add(nei)
                queue.append(nei)
    return order

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': [],
}

print(bfs(graph, 'A'))
```

### Why Learn Breadth‑First Search?

Breadth‑first search (BFS) explores a graph level by level.  It is ideal for
finding the shortest path in unweighted networks, such as finding the
shortest route in a maze or between web pages.  BFS is also used to detect
cycles, test connectivity and build spanning trees.  In networking, BFS
helps in broadcasting messages and in peer‑to‑peer networks.
